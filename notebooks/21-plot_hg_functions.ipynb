```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sys
if '..' not in sys.path: sys.path.append('..'); sys.path.append('../external')
    
from f_ring_util.f_ring import hg_func
```

```python
fig, axs = plt.subplots(2, 1, figsize=(15,15))
xrange = np.linspace(0., 180., 180)
ax = axs[0]
for g in np.arange(0, 0.03, 0.005):
    hg = hg_func((g, 1), xrange)
    ax.plot(xrange, hg, label=f'{g:.3f}')
ax.legend(ncol=3)
ax.set_xlim(0,180)
ax.set_yscale('log')
ax.set_ylabel('H-G Function')
ax = axs[1]
for g in np.arange(-.9, 0.01, 0.1):
    hg = hg_func((g, 1), xrange)
    ax.plot(xrange, hg, label=f'{g:.3f}')
ax.legend(ncol=3)
ax.set_xlim(0,180)
ax.set_yscale('log')
ax.set_xlabel('Phase Angle (°)')
ax.set_ylabel('H-G Function')
plt.suptitle('H-G Function with Various Parameters')
plt.tight_layout()
```
