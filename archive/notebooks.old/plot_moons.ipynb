```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
prometheus_data = pd.read_csv('../data_files/prometheus_historical.csv', parse_dates=['Date'], index_col='Date')
pandora_data = pd.read_csv('../data_files/pandora_historical.csv', parse_dates=['Date'], index_col='Date')
joint_data = prometheus_data.join(pandora_data, on='Date', how='inner')
joint_data.index = pd.to_datetime([x[:10] for x in joint_data.index])
```

```python
ax1 = plt.axes()
ax1.plot(joint_data.index, joint_data['Prometheus Distance'], '-', color='blue')
ax1.tick_params(axis='y', labelcolor='black')
ax1.set_ylabel('Prometheus Closest Approach (km)')
ax1.yaxis.label.set_color('blue')
ax2 = ax1.twinx()
ax2.plot(joint_data.index, joint_data['Pandora Distance'], '-', color='red')
ax2.tick_params(axis='y', labelcolor='red')
ax2.set_ylabel('Pandora Closest Approach (km)')
ax2.yaxis.label.set_color('red')
```
