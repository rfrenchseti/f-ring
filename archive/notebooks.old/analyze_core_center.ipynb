```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
obsdata = pd.read_csv('../data_files/ew_stats_139470_140170_140265_140965_0_core.csv', parse_dates=['Date'])
print('** SUMMARY STATISTICS **')
print('Unique observation names:', len(obsdata.groupby('Observation')))
print('Total 1-degree slices:', len(obsdata))
print('Starting date:', obsdata['Date'].min())
print('Ending date:', obsdata['Date'].max())
print('Time span:', obsdata['Date'].max()-obsdata['Date'].min())
```

```python
plt.plot(np.abs(np.cos(np.radians(obsdata['Mean Emission']))), obsdata['Core Offset Median'], 
         '.', ms=5)
plt.xlabel('$\mu=|\cos(e)|$')
plt.ylabel('Core Offset (km)')
_=plt.title('Core Offset vs. $\mu$ (Median)')
```

```python
plt.errorbar(np.abs(np.cos(np.radians(obsdata['Mean Emission']))), obsdata['Core Offset Mean'], 
             yerr=obsdata['Core Offset Std'], fmt='.', ms=10)
plt.xlabel('$\mu=|\cos(e)|$')
plt.ylabel('Core Offset (km)')
_=plt.title('Core Offset vs. $\mu$ (Mean)')
```

```python
plt.plot(np.abs(np.cos(np.radians(obsdata['Incidence']))), obsdata['Core Offset Median'], 
         '.')
plt.xlabel('$\mu_0=|\cos(i)|$')
plt.ylabel('Core Offset (km)')
_=plt.title('Core Offset vs. $\mu_0$ (Median)')
```

```python
plt.errorbar(np.abs(np.cos(np.radians(obsdata['Incidence']))), obsdata['Core Offset Mean'], 
             yerr=obsdata['Core Offset Std'], fmt='.')
plt.xlabel('$\mu_0=|\cos(i)|$')
plt.ylabel('Core Offset (km)')
_=plt.title('Core Offset vs. $\mu_0$ (Mean)')
```

```python
plt.plot(obsdata['Date'], obsdata['Core Offset Median'], 
         '.')
plt.xlabel('Date of Observation')
plt.ylabel('Core Offset (km)')
_=plt.title('Core Offset vs. Date (Median)')
```

```python
plt.errorbar(obsdata['Date'], obsdata['Core Offset Mean'], 
             yerr=obsdata['Core Offset Std'], fmt='.')
plt.xlabel('Date of Observation')
plt.ylabel('Core Offset (km)')
_=plt.title('Core Offset vs. Date (Mean)')
```
