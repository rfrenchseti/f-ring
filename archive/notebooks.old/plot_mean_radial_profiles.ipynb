```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
uvisdata = pd.read_csv('../data_files/uvis_aggregate_50.csv')
vimsdata = pd.read_csv('../data_files/vims_aggregate_50.csv')
ewdata0 = pd.read_csv('../data_files/ew_aggregate_tau0.csv')
ewdata = pd.read_csv('../data_files/ew_aggregate.csv')
```

```python
fig, ax1 = plt.subplots()
ax1.set_xlabel('Offset from Core (km)')
uvisplot, = ax1.plot(uvisdata['Radius'], uvisdata['Mean Optical Depth'], color='red', lw=4, alpha=0.5)
vimsplot, = ax1.plot(vimsdata['Radius'], vimsdata['Mean Optical Depth'], color='blue', lw=4, alpha=0.5)
ax1.tick_params(axis='y', labelcolor='black')
ax1.set_ylabel('Mean Tau')
ax1.legend()
ax2 = ax1.twinx()
issplot, = ax2.plot(ewdata['Radius'], ewdata['Mean Normal EW3ZPN'], color='grey', lw=2)
issplot0, = ax2.plot(ewdata['Radius'], ewdata0['Mean Normal EW3ZPN'], color='black', lw=2)
ax2.tick_params(axis='y', labelcolor='green')
ax2.set_ylabel('Mean PN I/F (km)')
ax2.yaxis.label.set_color('green')
plt.xlim(-500,500)
ax2.legend([uvisplot, vimsplot, issplot, issplot0], ['UVIS', 'VIMS', 'ISS Tau 0.042', 'ISS No Tau'])
_=plt.title('Mean Occultation vs. Mosaic Comparison')
print(np.max(uvisdata['Mean Optical Depth']), np.max(vimsdata['Mean Optical Depth']))
```
